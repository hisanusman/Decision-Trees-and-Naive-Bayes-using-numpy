## Navive bayes class (without scikit-learn)

```python
class naive_bayes():

  def __init__(self, user_outlook, user_temperature, user_humidity, user_wind):
    self.Outlook = user_outlook
    self.Temperature = user_temperature
    self.Humidity = user_humidity
    self.Wind = user_wind


  def unique(self, df):
    dict_uniques = {'Outlook': df['Outlook'].unique(),
                    'Temperature': df['Temperature'].unique(),
                    'Humidity': df['Humidity'].unique(),
                    'Wind': df['Wind'].unique()}

    return dict_uniques


  def probability(self, df):
    count_yes = (df['Play Tennis'] == 'Yes').sum()
    count_no = (df['Play Tennis'] == 'No').sum()

    prob_yes = count_yes / len(df)
    prob_no = count_no / len(df)

    count_user_outlook_yes = ((df['Outlook'] == self.Outlook) & (df['Play Tennis'] == 'Yes')).sum()
    count_user_temperature_yes = ((df['Temperature'] == self.Temperature) & (df['Play Tennis'] == 'Yes')).sum()
    count_user_humidity_yes = ((df['Humidity'] == self.Humidity) & (df['Play Tennis'] == 'Yes')).sum()
    count_user_wind_yes = ((df['Wind'] == self.Wind) & (df['Play Tennis'] == 'Yes')).sum()

    count_user_outlook_no = ((df['Outlook'] == self.Outlook) & (df['Play Tennis'] == 'No')).sum()
    count_user_temperature_no = ((df['Temperature'] == self.Temperature) & (df['Play Tennis'] == 'No')).sum()
    count_user_humidity_no = ((df['Humidity'] == self.Humidity) & (df['Play Tennis'] == 'No')).sum()
    count_user_wind_no = ((df['Wind'] == self.Wind) & (df['Play Tennis'] == 'No')).sum()

    prob_user_outlook_yes = count_user_outlook_yes / count_yes
    prob_user_temperature_yes = count_user_temperature_yes / count_yes
    prob_user_humidity_yes = count_user_humidity_yes / count_yes
    prob_user_wind_yes = count_user_wind_yes / count_yes

    prob_user_outlook_no = count_user_outlook_no / count_no
    prob_user_temperature_no = count_user_temperature_no / count_no
    prob_user_humidity_no = count_user_humidity_no / count_no
    prob_user_wind_no = count_user_wind_no / count_no

    total_prob_yes = prob_yes * prob_user_outlook_yes * prob_user_temperature_yes * prob_user_humidity_yes * prob_user_wind_yes
    total_prob_no = prob_no * prob_user_outlook_no * prob_user_temperature_no * prob_user_humidity_no * prob_user_wind_no

    if total_prob_yes > total_prob_no:
        print("Go play\n", "Probability is:", total_prob_yes)
    else:
        print("Don't play\n", "Probability is:", total_prob_no)


  def calculate_accuracy(self, df):
        correct_predictions = 0
        total_predictions = 0

        for index, row in df.iterrows():
            actual_label = row['Play Tennis']
            predicted_label = self.probability(df)

            if predicted_label == actual_label:
                correct_predictions += 1
            total_predictions += 1

        accuracy = correct_predictions / total_predictions
        return accuracy
```

## Getting Predictions:

```python
check = naive_bayes(user_outlook, user_temperature, user_humidity, user_wind)
```
